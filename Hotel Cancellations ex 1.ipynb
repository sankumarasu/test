```python
import pandas as pd
```

```python
df = pd.read_csv('../data/H1.csv')
```

```python
df.head()
```

```python
df['YearWeek'] = df['ArrivalDateYear'].map(str) + '-' + df['ArrivalDateWeekNumber'].map(str)
```

```python
df.head()
```

```python
df.groupby(['YearWeek']).agg({'IsCanceled':'sum'}).plot()
```

```python
canceled_ts = df.groupby(['YearWeek']).agg({'IsCanceled':'sum'})
canceled_ts.head()
```
